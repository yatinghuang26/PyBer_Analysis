# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Import NumPy and the stats module from SciPy.
import numpy as np
import scipy.stats as sts
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_ride_count = np.sum(urban_cities_df.groupby(["city"]).count()["ride_id"])

suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_ride_count = np.sum(suburban_cities_df.groupby(["city"]).count()["ride_id"])

rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_ride_count = np.sum(rural_cities_df.groupby(["city"]).count()["ride_id"])
```

```python
# 2. Get the total drivers for each city type
urban_drivers_count = np.sum(urban_cities_df.groupby(['city']).sum()["driver_count"])
suburban_drivers_count = np.sum(suburban_cities_df.groupby(['city']).sum()["driver_count"])
rural_drivers_count = np.sum(rural_cities_df.groupby(['city']).sum()["driver_count"])
```

```python
#  3. Get the total amount of fares for each city type
urban_total_fares = np.sum(urban_cities_df.groupby(['city']).sum()["fare"])
suburban_total_fares = np.sum(suburban_cities_df.groupby(['city']).sum()["fare"])
rural_total_fares = np.sum(rural_cities_df.groupby(['city']).sum()["fare"])
```

```python
#  4. Get the average fare per ride for each city type. 
urban_avg_fare_ride = urban_total_fares / urban_ride_count
suburban_avg_fare_ride = suburban_total_fares / suburban_ride_count
rural_avg_fare_ride = rural_total_fares / rural_ride_count
```

```python
# 5. Get the average fare per driver for each city type. 
urban_avg_fare_driver = urban_total_fares / urban_drivers_count
suburban_avg_fare_driver = suburban_total_fares / suburban_drivers_count
rural_avg_fare_driver = rural_total_fares / rural_drivers_count
```

```python
#  6. Create a PyBer summary DataFrame. 
PyBer_summary_df = pd.DataFrame(
          {"type": ['rural', 'suburban', 'urban'], 
          "Total Rides": [rural_ride_count, suburban_ride_count, urban_ride_count],
          "Total Drivers": [rural_drivers_count, suburban_drivers_count, urban_drivers_count],
          "Total Fares": [rural_total_fares, suburban_total_fares, urban_total_fares],
          "Average Fare per Ride": [rural_avg_fare_ride, suburban_avg_fare_ride, urban_avg_fare_ride],
          "Average Fare per Driver": [rural_avg_fare_driver, suburban_avg_fare_driver, urban_avg_fare_driver]})
PyBer_summary_df = PyBer_summary_df.set_index("type")
PyBer_summary_df

```

```python
#  7. Cleaning up the DataFrame. Delete the index name
PyBer_summary_df.index.name = None
PyBer_summary_df
```

```python
#  8. Format the columns.
PyBer_summary_df["Total Rides"] = PyBer_summary_df["Total Rides"].map("{:,}".format)
PyBer_summary_df["Total Drivers"] = PyBer_summary_df["Total Drivers"].map("{:,}".format)
PyBer_summary_df["Total Fares"] = PyBer_summary_df["Total Fares"].map("${:,.2f}".format)
PyBer_summary_df["Average Fare per Ride"] = PyBer_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
PyBer_summary_df["Average Fare per Driver"] = PyBer_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
PyBer_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_merged_df = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
pyber_data_merged_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_data_merged_df = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
pyber_data_merged_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_data_merged_df = pyber_data_merged_df.reset_index()
pyber_data_merged_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_merged_df_pivot = pyber_data_merged_df.pivot(index = "date", columns = "type", values = "fare")
pyber_data_merged_df_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_data_merged_df_pivot = pyber_data_merged_df_pivot.reset_index()
pyber_data_merged_df_pivot = pyber_data_merged_df_pivot.loc[(pyber_data_merged_df_pivot["date"]>= '2019-01-01') & (pyber_data_merged_df_pivot["date"]<= '2019-04-29')]
pyber_data_merged_df_pivot = pyber_data_merged_df.pivot(index = "date", columns = "type")
pyber_data_merged_df_pivot
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_data_merged_df_pivot.index = pd.to_datetime(pyber_data_merged_df_pivot.index)
pyber_data_merged_df_pivot
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_data_merged_df_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_data_merged_df_new_week = pyber_data_merged_df_pivot.resample("W").sum()
pyber_data_merged_df_new_week.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fig, ax = plt.subplots()
pyber_data_merged_df_new_week.plot(kind = "line", ax=ax)
ax.legend(["Rural", "Suburban", "Urban"])
ax.set_ylabel("Fare($)")
ax.set_title("Total Fare by City Type")
plt.savefig("analysis/PyBer_fare_summary.png")

```
